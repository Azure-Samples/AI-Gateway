# Clean Up Resources

This notebook removes the resource group created for the `ai-foundry-hosted-agents-custom-framework` lab.

```python
import shutil
import subprocess
import sys

_az = (
    shutil.which("az")
    or shutil.which("az.cmd")
    or r"C:\Program Files\Microsoft SDKs\Azure\CLI2\wbin\az.CMD"
 )

if not _az:
    raise FileNotFoundError(
        "Azure CLI executable was not found. Install Azure CLI or add it to PATH."
    )

def run(cmd):
    if cmd and cmd[0] == "az":
        cmd = [_az] + cmd[1:]
    print("\n$ " + " ".join(cmd))
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.stdout:
        print(result.stdout)
    if result.returncode != 0:
        if result.stderr:
            print(result.stderr)
        raise RuntimeError(f"Command failed: {result.returncode}")

subscription_id = "baba41cf-c01d-4a55-b6c5-ca494b802be5"
resource_group = "rg-aigw-hosted-custom-framework"
```

```python
# Delete lab resources
run(["az", "account", "set", "--subscription", subscription_id])
run(["az", "group", "delete", "--name", resource_group, "--yes", "--no-wait"])
print("Deletion started. Use az group show to verify completion.")
```
