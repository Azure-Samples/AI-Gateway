### Delete the Entra security group

We also remove the approved-users security group that was created for the lab.

```python
allowed_group_name = f"{deployment_name}-users"

print(f"Searching for group with display name {allowed_group_name}...")
output = utils.run(f"az ad group list --filter \"displayName eq '{allowed_group_name}'\"", "Retrieved groups", "Failed to list groups")
if output.success and output.json_data:
    group_id = output.json_data[0]['id']
    utils.run(f"az ad group delete --group {group_id}", "Deleted the security group", "Failed to delete the group")
```
